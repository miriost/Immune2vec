### DATA FILTERING - Filter hexagons that originated from only one subject

```python
from bokeh.util.hex import cartesian_to_axial

q0, r0 = cartesian_to_axial(TSNE_50K['dim1'],TSNE_50K['dim2'], size=3, orientation="pointytop")
TSNE_50K['center']=[(q,r) for (q,r) in zip(q0, r0)]

df1 = TSNE_50K.groupby(['center','SUBJECT']).size().reset_index(name='counts')
df2 = TSNE_50K.groupby(['center']).size().reset_index(name='counts')
for index, row in df2.iterrows():
    if len(df1[df1.center==row['center']].SUBJECT.unique())>1:
        print(str(row['center'])+ ' SUBJECTS: '+ str(df1[df1.center==row['center']].SUBJECT.unique()))
    else:
        print('removing cell: ' + str(row['center']) + ' FROM SUBJECT: ' + str(df1[df1.center==row['center']].SUBJECT.unique()))
        TSNE_50K = TSNE_50K[TSNE_50K.center != row['center']]
                
```

```python
print('Total number of filtered sequnces: ' + str(50000-len(TSNE_50K)) + '/50000')
print('Which are: ' + str(100-100*len(TSNE_50K)/50000) + '%')
```

## syntetic example - remove later

```python
import numpy as np

from bokeh.io import output_file, show
from bokeh.models import HoverTool, ColorBar
from bokeh.plotting import figure

n = 500
x = 2 + 2*np.random.standard_normal(n)
y = 2 + 2*np.random.standard_normal(n)

p = figure(title="Hexbin for 500 points", match_aspect=True,
           tools="wheel_zoom,reset", background_fill_color='#440154')
p.grid.visible = False

r, bins = p.hexbin(x, y, size=4, hover_color="pink", hover_alpha=0.8)

p.circle(x, y, color="white", size=1)

hover = HoverTool(tooltips=[("count", "@c"), ("(q,r)", "(@q, @r)")],
                  mode="mouse", point_policy="follow_mouse", renderers=[r])

p.add_tools(hover)
color_bar = ColorBar()
output_file("hexbin.html", mode="inline")
p.add_layout(color_bar, 'right')
show(p)
```

```python
from bokeh.util.hex import cartesian_to_axial

#def filter_haxagons()

q0, r0 = cartesian_to_axial(x, y, size=4, orientation="pointytop")
df = pd.DataFrame(dict(center=[(q,r) for (q,r) in zip(q0, r0)], 
                       condition = np.random.randint(5, size = len(r0)), indexes = range(len(x)),x=x,y=y))
df1 = df.groupby(['center','condition']).size().reset_index(name='counts')
df2 = df.groupby(['center']).size().reset_index(name='counts')
for index, row in df2.iterrows():
    if len(df1[df1.center==row['center']].condition.unique())>1:
        print(str(row['center'])+ str(df1[df1.center==row['center']].condition.unique()))
    else:
        print('removing cell: ' + str(row['center']))
        df = df[df.center != row['center']]
```

```python
p = figure(title="Filtered Hexbin", match_aspect=True,
           tools="wheel_zoom,reset", background_fill_color='#440154')
p.grid.visible = False

r, bins = p.hexbin(df['x'], df['y'], size=4, hover_color="pink", hover_alpha=0.8)

p.circle(x, y, color="white", size=1)

hover = HoverTool(tooltips=[("count", "@c"), ("(q,r)", "(@q, @r)")],
                  mode="mouse", point_policy="follow_mouse", renderers=[r])

p.add_tools(hover)
color_bar = ColorBar()
output_file("hexbin.html", mode="inline")
p.add_layout(color_bar, 'right')
show(p)
```
